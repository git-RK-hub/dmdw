Sometimes negative values are read as positive, to crosscheck we look at 'Quantity' column. And we found that there are negative values indeed. Fortunately no other column has negative values where they should not be.
In the next few cells we clean them up.

```python
df['Quantity'].min()
```

```python
df['UnitPrice'].min()
```

```python
zx=df['Quantity']<=0
rem=[i for i in range(len(zx)) if zx[i]]
```

```python
len(rem)
```

```python
df=df.drop(rem, axis=0)
df=df.reset_index()
df=df.drop('index', axis=1)
df
```

Additional Columns are created, for better analysis.

```python
df['Total Amount Spent']= df['Quantity'] * df['UnitPrice']
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'], format='%d-%m-%Y %H:%M')

final = df['InvoiceDate'].max()
df['Last_transact'] = final - df['InvoiceDate']
df['Last_transact']=df['Last_transact'].dt.days
```

```python
df.rename(columns = {'Last_transact':'Last_Shop'}, inplace = True)
df
```

```python
rem=['InvoiceNo', 'Description']
```

```python
df=df.drop(rem, axis=1)
df
```

```python
df['CustomerID'].dtype
```

```python
df['CustomerID']=df['CustomerID'].astype(int)
df['CustomerID'].dtype
```

A new column is created which will help us when we'll do case by case analyses at the later part.

```python
cus=df.CustomerID.unique()
len(cus)
```

```python
o=np.ones((397924, 1), dtype='int8')
df['Cust_value']=o
```

```python
country=df.groupby('Country').sum()
country
```

```python
con=df['Country'].unique()
con.sort()
```

```python
no_cus={}
qua={}
toamt={}
for i,c in enumerate(con):
    no_cus[c]=country['Cust_value'][i]
    qua[c]=country['Quantity'][i]
    toamt[c]=country['Total Amount Spent'][i]
```

```python
list_of_col = list(zip(con, no_cus.values(), toamt.values(), qua.values()))
pro1_df = pd.DataFrame(list_of_col, columns = ['Country', 'No. of Customers', 'Trade_Amount', 'Quantity'])
```

```python
pro1_df
```

```python
sc=df['StockCode'].unique()
sc.sort()
```

```python
stocks=df.groupby('StockCode').sum()
stocks
```

```python
stocks=stocks.drop(['CustomerID', 'Total Amount Spent', 'Last_Shop'], axis=1)
stocks
```

```python
stocks['UnitPrice']=(stocks['UnitPrice']/stocks['Cust_value'])
```

```python
unit_price={}
stock_qua={}
nof_trans={}
for i,c in enumerate(sc):
    unit_price[c]=stocks['UnitPrice'][i]
    stock_qua[c]=stocks['Quantity'][i]
    nof_trans[c]=stocks['Cust_value'][i]
```

```python
list_of_col = list(zip(sc, unit_price.values(), nof_trans.values(), stock_qua.values()))
pro2_df = pd.DataFrame(list_of_col, columns = ['StockCode', 'UnitPrice', 'No. of Transactions', 'Quantity'])
pro2_df['Revenue']=pro2_df['UnitPrice']*pro2_df['Quantity']
```

# Now the graphs can be created. We have created 2 dataframes from initial csv file. 

```python
data=pro1_df.sort_values(by='Trade_Amount', ascending=True)
fig = plt.figure(figsize = (10, 5))
x=data['Country'][:5]
y=data['Trade_Amount'][:5]
# creating the bar plot 
plt.bar(data['Country'][:5], data['Trade_Amount'][:5], color='r', width=0.4) 
for index, value in enumerate(y):
    plt.text(index,value, str(round(value)))
plt.xlabel("Country") 
plt.ylabel("Trade_Value in $") 
plt.title("Least Trading Country") 
plt.show() 
```

```python
data=pro2_df.sort_values(by='UnitPrice', ascending=False)
fig = plt.figure(figsize = (10, 5)) 
x=data['StockCode'][:5]
y=data['UnitPrice'][:5]
# creating the bar plot 
plt.bar(data['StockCode'][:5], data['UnitPrice'][:5], color='r', width=0.4) 
for index, value in enumerate(y):
    plt.text(index,value, str(round(value)))
plt.xlabel("Stock Code") 
plt.ylabel("UnitPrice in $") 
plt.title("Most Expensive Stocks") 
plt.show() 
```

```python
fig = plt.figure(figsize = (10, 5)) 
y=data['Revenue'][:5]
# creating the bar plot 
plt.bar(x, y, color='r', width=0.4) 
for index, value in enumerate(y):
    plt.text(index,value, str(round(value)))
plt.xlabel("Stock Code") 
plt.ylabel("Revenue in $") 
plt.title("Revenue of most expensive stocks") 
plt.show() 
```

```python
data=pro2_df.sort_values(by='No. of Transactions', ascending=False)
fig = plt.figure(figsize = (10, 5)) 
x=data['StockCode'][:5]
y=data['No. of Transactions'][:5]
# creating the bar plot 
plt.bar(x, y, color='r', width=0.4)
for index, value in enumerate(y):
    plt.text(index,value, str(round(value)))
plt.xlabel("Stock Code") 
plt.ylabel("Transactions") 
plt.title("Most Traded Stocks") 
plt.show() 
```

```python
def func(pct, allvalues): 
    absolute = int(pct / 100.*np.sum(allvalues)) 
    return "{:.1f}%".format(pct, absolute)
```

```python
data=pro1_df.sort_values(by='No. of Customers', ascending=False)
labels=list(data['Country'][:5])
labels.append('others')
labels[3]='Ireland'
values=list(data['No. of Customers'][:5])
values.append(data['No. of Customers'][5:].sum())
values
```

```python
fig_cent = plt.figure(figsize =(10, 7)) 
plt.pie(values, labels = labels, autopct = lambda pct: func(pct, values))
plt.legend(title ="Country", loc ="best")
plt.title('Market Size of Countries')
plt.show()
```
